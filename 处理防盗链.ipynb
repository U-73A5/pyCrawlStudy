```python
import requests

url = 'https://www.pearvideo.com/video_1750269'
contId = url.split('_')[-1]
videoStatus = f'https://www.pearvideo.com/videoStatus.jsp?contId=1750269&mrd=0.5896963326540208'

headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.74 Safari/537.36 Edg/99.0.1150.55',
  'Referer': url # 防盗链
}

resp = requests.get(videoStatus, headers=headers)
resp.close()
srcURL = resp.json()['videoInfo']['videos']['srcUrl']
systemTime = resp.json()['systemTime']

srcURL = srcURL.replace(systemTime, f'cont-{contId}')
resp = requests.get(srcURL)
resp.close()
with open('video.mp4', 'wb') as f:
  f.write(resp.content)
```
