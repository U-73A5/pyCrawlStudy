```python
import requests
from lxml import etree

url = 'https://haerbin.zbj.com/search/f/?kw=saas'

resp = requests.get(url)
resp.close()

tree = etree.HTML(resp.text)
divs = tree.xpath('/html/body/div[6]/div/div/div[2]/div[5]/div[1]/div')
for div in divs:
  print(div.xpath('./div/div/a[1]/div[1]/p/text()')[1], '\n',
    'saas'.join(div.xpath('./div/div/a[2]/div[2]/div[2]/p/text()')), '\n',
    div.xpath('./div/div/a[2]/div[2]/div[1]/span[1]/text()')[0][1:],
    sep='', end='')
```
