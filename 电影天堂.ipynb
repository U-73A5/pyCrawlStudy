```python
import requests
import re
from urllib import parse

domain = 'https://dytt89.com/'
resp = requests.get(domain) # verify=False去掉安全验证
resp.encoding = 'gb2312' # 指定字符集
resp.close()

objUl = re.compile(r'2022必看热片.*?<ul>(?P<ul>.*?)</ul>', re.S)
resultUl = objUl.finditer(resp.text)
for it in resultUl:
  ul = it.group('ul')
objA = re.compile(r'<a href=\'(?P<href>.*?)\'', re.S)
resultA = objA.finditer(ul)
url = []
for it in resultA:
  url.append(parse.urljoin(domain, it.group('href')))

objP = re.compile(r'◎片　　名(?P<name>.*?)<br'
                  r'.*?<td.*?<a href="(?P<downloadT>.*?)">'
                  , re.S)
for herf in url:
  res = requests.get(herf)
  res.encoding = 'gb2312'
  resultT = objP.search(res.text)
  print(resultT.group('name'), resultT.group('downloadT'))
  res.close()
```
