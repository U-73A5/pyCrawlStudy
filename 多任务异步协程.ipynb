```python
# import time
# print(1)
# time.sleep(3) # 当前线程处于阻塞， cpu停止为线程工作 # 一般 IO操作
# print(2)

# 阻塞操作 sleep input request.xxx()

# 单线程下：
#   协程：选择性切换到其他任务，使 cup持续为自己工作
#   多任务异步操作

import time
import asyncio

async def func():
  print('async')
  # time.sleep(3) # 同步操作，异步中断
  await asyncio.sleep(3) # 异步操作，await挂起
async def func2():
  # time.sleep(3)
  await asyncio.sleep(3)
  print(233)
async def func3():
  # time.sleep(2)
  await asyncio.sleep(2)
  print(233)

# func() # 返回协程对象
# asyncio.run(func())

# tasks = [func(), func2(), func3()]
# asyncio.run(asyncio.wait(tasks)) # can't run in Jupyter

async def main():
  # await func() # await 在 async下，一般在协程对象前
  # await func2()
  tasks = [
    asyncio.create_task(func()), # py3.8 后
    asyncio.create_task(func2()),
    asyncio.create_task(func3())
  ]
  await asyncio.wait(tasks)

asyncio.run(main())
```
