```python
import os
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.umeitu.com/katongdongman/katongtupian/'
resp = requests.get(url)
resp.close()
resp.encoding = 'utf-8'

page = bs(resp.text, 'html.parser')
div = page.find('div', attrs={'class': 'TypeList'})
imgs = div.find_all('img')
for img in imgs:
  src = img.get('src') # img.attrs['src']
  src = src.split('small')
  src = src[0] + src[1]
  # print(src)
  rep = requests.get(src)
  rep.close()
  with open(os.path.join('img', src.split('/')[-1]), 'wb') as f:
    f.write(rep.content)
```
