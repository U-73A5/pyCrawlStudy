#### GET headers text close

```python
import requests

url = 'https://cn.bing.com/search?q=matlab+multi-label+classification'

headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.74 Safari/537.36 Edg/99.0.1150.46',
}

resp = requests.get(url, headers=headers) # GET获取网页
resp.close() # 关掉resp

print(resp) # ouput：<Response [200]>
print(resp.text) #页面源代码
```

#### POST data json

```python
import requests

url = 'https://fanyi.baidu.com/sug'

dat = {
  'kw': 'dog'
}

resp = requests.post(url, data=dat)
resp.close()

print(resp)
print(resp.json()) # 将内容处理为json -> dict
```

#### GET-params response-attr

```python
import requests

url = 'https://movie.douban.com/j/chart/top_list'

head = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.74 Safari/537.36 Edg/99.0.1150.46'
}
param = {
  'type': '11',
  'interval_id': '100:90',
  'action': '',
  'start': '0',
  'limit': '20',
}

resp = requests.get(url, headers=head, params=param)
resp.close()

print(resp.request.headers) # 响应--响应发起对象--发起对象的attr
print(resp.json()) # 客户端渲染返回数据直接转换json
```
