```python
from threading import Thread

def func(a):
  for i in range(10000):
    print(f'{a}func{i}')

t = Thread(target=func, args=('123',)) # 创建线程
t.start() # 给线程一个开始信号

for i in range(10000):
  print(f'a{i}')


# class MyThread(Thread):
#   def __init__(self, name):
#     self.name = name
#   def run(self):
#     for i in range(10000):
#       print(f'{self.name}func{i}')
```
