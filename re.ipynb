```python
import re

# 匹配字符串中所有符合正则的内容，返回list
lst = re.findall(r'\d+', 'my934997459you123456')
print(lst)

# 匹配字符串中所有内容，返回迭代器
# 获取内容用group，效率高于list
it = re.finditer(r'\d+', 'my934997459you123456')
print(it)
for i in it:
  print(i.group())
```

```python
import re

# 全文匹配，找到一个结果就返回match
s = re.search(r'\d+', 'my934997459you123456')
print(s.group())

# 从头匹配，返回match
m = re.match(r'\d', 'my934997459you123456')
print(m)
```

```python
import re

# 预加载正则表达式
obj = re.compile(r'\d+')
lst = obj.findall('my934997459you123456')
print(lst)
```

```python
import re

s = '''
<div class='cxk'><span id='4'>sgdd</span></id>
<div class='chuanshanjia'><span id='4'>jitang</span></id>
<div class='huaqing'><span id='4'>maigua</span></id>
<div class='chenglong'><span id='1'>awooo</span></id>
<div class='jiege'><span id='3'>gcs</span></id>
'''

# (?P<分组名字>正则) 单独从正则匹配的内容中进一步提取内容
# 参数 flag额外的规则，re.S使.能够匹配换行符
obj = re.compile(r'''<div class='(?P<name>.*?)'><span id='(?P<id>\d+)'>(?P<action>.*?)</span></id>''', re.S)
result = obj.finditer(s)
for it in result:
  print(it.group('id'))
  print(it.group('name'))
  print(it.group('action'))
```
