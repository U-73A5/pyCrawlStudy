```python
from lxml import etree

xml = '''
<book>
  <id>1</id>
  <name>u73a5</name>
  <nicks>
    <nick>qkndlxz</nick>
    <nick>byjq</nick>
    <truth>
      <nick>ljh</nick>
      <nick>lxz</nick>
    </truth>
    <nonuse>
      <nick>lzy</nick>
    </nonuse>
  </nicks>
</book>
'''

tree = etree.XML(xml) # etree.parse() 加载文件
print(tree.xpath('/book')) # / 根节点
print(tree.xpath('/book/name/text()')) # text() 获取文本
print(tree.xpath('/book/nicks/nick/text()'))
print(tree.xpath('/book/nicks//nick/text()')) # // 获取所有后辈节点
print(tree.xpath('/book/nicks/*/nick/text()')) # * 通配符，任意节点
```

```python
import requests
from lxml import etree

url = 'https://www.sojson.com/web/use.html'
head = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.74 Safari/537.36 Edg/99.0.1150.55'
}
resp = requests.get(url, headers=head)
resp.close()

tree = etree.HTML(resp.text)
print(tree.xpath('/html/body/div[1]/div[5]/div[2]/div[1]/fieldset[1]//h1/text()')) # xpath [] 第一个从 1 开始
print(tree.xpath('''/html/body/div[@class='layout']/div[@id='so_box']/div[2]/div[1]/
fieldset[@class='layui-elem-field layui-field-title site-title']//h1/text()''')) # xpath 节点属性 [@属性=属性值]
print(tree.xpath('''/html/body/div[@class='layout']/div[@id='so_box']/div[2]/div[1]/
fieldset[@class='layui-elem-field layui-field-title site-title']//h1/@id''')) # 获取属性值 @
```
